```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data hasil saringan
stories = ['1 Lantai', '2 Lantai', '3 Lantai', '4 Lantai']
jumlah_rumah = [43, 33, 2, 1]

# Mengatur gaya visualisasi
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))

# Membuat Countplot / Bar Chart
ax = sns.barplot(x=stories, y=jumlah_rumah, palette="viridis")

# Menambahkan judul dan label kapak
plt.title("Sebaran Jumlah Rumah Berdasarkan Jumlah Lantai\n(Area > Rata-rata & Price < Rata-rata)", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Jumlah Lantai (Stories)", fontsize=12)
plt.ylabel("Jumlah Rumah", fontsize=12)

# Menambahkan angka di atas setiap batang grafik
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='baseline', fontsize=11, color='black', weight='bold',
                xytext=(0, 5), textcoords='offset points')

# Menampilkan grafik
plt.tight_layout()
plt.show()
```
