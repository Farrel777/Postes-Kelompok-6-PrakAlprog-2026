```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data
df = pd.read_csv("Kelas C_Housing.csv")

# Menghitung standar deviasi price berdasarkan stories
std_price = df.groupby('stories')['price'].std()

# Menampilkan hasil
print(std_price)

# Membuat bar chart
std_price.plot(kind='bar')
plt.title('Standar Deviasi Harga Rumah berdasarkan Jumlah Lantai')
plt.xlabel('Stories (Jumlah Lantai)')
plt.ylabel('Standar Deviasi Price')
plt.show()
```
