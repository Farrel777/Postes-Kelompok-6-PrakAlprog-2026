GABUNGAN SEMUA KATEGORI

```python
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# BACA DATA
# =========================
df = pd.read_csv("Kelas C_Housing.csv")

# =========================
# DATA GRAFIK 1
# =========================
std_price = df.groupby('stories')['price'].std()

# =========================
# DATA GRAFIK 2
# =========================
stories = ['1 Lantai', '2 Lantai', '3 Lantai', '4 Lantai']
jumlah_rumah = [43, 33, 2, 1]

# =========================
# DATA GRAFIK 3
# =========================
df['total_rooms'] = df['bedrooms'] + df['bathrooms']

# =========================
# DATA GRAFIK 4
# =========================
harga = df['price']

q1 = harga.quantile(0.3333)
q2 = harga.quantile(0.6667)

def kategori_harga(x):
    if x <= q1:
        return 'Murah'
    elif x <= q2:
        return 'Sedang'
    else:
        return 'Mewah'

df['Kategori'] = harga.apply(kategori_harga)

distribusi = df['Kategori'].value_counts()
distribusi = distribusi.reindex(['Murah', 'Sedang', 'Mewah'])

# =========================
# GABUNGAN 2x2
# =========================
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# -------------------------
# Grafik 1
# -------------------------
std_price.plot(
    kind='bar',
    ax=axes[0,0],
    color='skyblue'
)

axes[0,0].set_title(
    'Standar Deviasi Harga Rumah\nBerdasarkan Jumlah Lantai'
)
axes[0,0].set_xlabel('Stories')
axes[0,0].set_ylabel('Std Price')

# -------------------------
# Grafik 2
# -------------------------
bars = axes[0,1].bar(
    stories,
    jumlah_rumah,
    color='lightgreen'
)

axes[0,1].set_title(
    'Sebaran Jumlah Rumah\n(Area > Rata-rata & Price < Rata-rata)'
)
axes[0,1].set_xlabel('Jumlah Lantai')
axes[0,1].set_ylabel('Jumlah Rumah')

for bar in bars:
    h = bar.get_height()
    axes[0,1].text(
        bar.get_x() + bar.get_width()/2,
        h,
        str(int(h)),
        ha='center',
        va='bottom'
    )

# -------------------------
# Grafik 3
# -------------------------
axes[1,0].scatter(
    df['total_rooms'],
    df['price'],
    alpha=0.7
)

axes[1,0].set_title(
    'Pengaruh Total Ruangan\nterhadap Harga Rumah'
)
axes[1,0].set_xlabel('Total Rooms')
axes[1,0].set_ylabel('Price')

# -------------------------
# Grafik 4
# -------------------------
colors = ['#66c2a5', '#ffd92f', '#fc8d62']

axes[1,1].pie(
    distribusi,
    labels=distribusi.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    explode=(0.03, 0.03, 0.03)
)

axes[1,1].set_title(
    'Distribusi Harga Rumah\n(Berdasarkan Kuantil)'
)

# =========================
# JUDUL UTAMA
# =========================
fig.suptitle(
    'Dashboard Analisis Data Housing',
    fontsize=16,
    fontweight='bold'
)

plt.tight_layout()
plt.subplots_adjust(top=0.90)
plt.show()
```
