```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data
df = pd.read_csv("Kelas C_Housing.csv")

# Membuat variabel total ruangan
df['total_rooms'] = df['bedrooms'] + df['bathrooms']

# Scatter plot
plt.scatter(df['total_rooms'], df['price'])
plt.xlabel('Total Rooms (Bedrooms + Bathrooms)')
plt.ylabel('Price')
plt.title('Pengaruh Total Ruangan terhadap Harga Rumah')
plt.show()
```
